```python
import pandas as pd
df = pd.read_csv("Downloads/archive (1)/Housing.csv")
X = df[["area", "bedrooms", "bathrooms"]]
Y = df["price"]

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 2)

from sklearn.linear_model import SGDRegressor
# Instantiate and train the SGD Regressor
sgd = SGDRegressor()
sgd.fit(X_train, Y_train)

# Make predictions
result_sgd = sgd.predict(X_test)

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
import matplotlib.pyplot as plt

# Calculate metrics
mse_sgd = mean_squared_error(Y_test, result_sgd)
rmse_sgd = np.sqrt(mse_sgd)
mae_sgd = mean_absolute_error(Y_test, result_sgd)
r2_sgd = r2_score(Y_test, result_sgd)

# Print metrics
print(f"Root Mean Squared Error (RMSE): {rmse_sgd}")
print(f"Mean Absolute Error (MAE): {mae_sgd}")
print(f"R^2 Score: {r2_sgd}")

# Plot predictions vs. true values
plt.figure(figsize=(10, 6))
plt.scatter(Y_test, result_sgd)
plt.xlabel("True Values")
plt.ylabel("Predictions")
plt.title("True Values vs Predictions (SGD Regression)")
plt.plot([min(Y_test), max(Y_test)], [min(Y_test), max(Y_test)], 'r--', lw=2)  # Perfect prediction line
plt.show()

# Print coefficients and intercept
print("Coefficients:", sgd.coef_)
```
