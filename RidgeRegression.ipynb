```python
import pandas as pd
df = pd.read_csv("Downloads/archive (1)/Housing.csv")
X = df[["area", "bedrooms", "bathrooms"]]
Y = df["price"]

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 2)

from sklearn.linear_model import Ridge
r = Ridge()
r = r.fit(X_train, Y_train)
result = r.predict(X_test)

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
import matplotlib.pyplot as plt

# Calculate metrics
mse = mean_squared_error(Y_test, result)
rmse = np.sqrt(mse)
mae = mean_absolute_error(Y_test, result)
r2 = r2_score(Y_test, result)

# Print metrics
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"R^2 Score: {r2}")

# Plot predictions vs. true values
plt.figure(figsize=(10, 6))
plt.scatter(Y_test, result)
plt.xlabel("True Values")
plt.ylabel("Predictions")
plt.title("True Values vs Predictions (Ridge Regression)")
plt.plot([min(Y_test), max(Y_test)], [min(Y_test), max(Y_test)], 'r--', lw=2)  # Perfect prediction line
plt.show()


# Print coefficients 
print("Coefficients:", r.coef_)
```
